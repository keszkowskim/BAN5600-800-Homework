```python
from google.colab import auth
import gspread
from google.auth import compute_engine
from google.auth import default

# Authenticate the user
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

print("Successfully authenticated!")
```

Excercise

```python
# --- 1. Create or Open a Sheet ---
# Replace 'My Commission Report' with whatever name you want
sh = gc.create('Commission Report')
worksheet = sh.get_worksheet(0)

# --- 2. Define our Data (Chapters 1-3 style) ---
# Header Row
headers = ["Sales Amount", "Region", "Final Commission"]
worksheet.append_row(headers)

```

```python
# Sample Data to write
sales_data = [
    [1200.00, "E", 98.00],
    [5500.00, "W", 355.00],
    [800.00, "C", 16.00]
]

# --- 3. Loop through and write ---
print("Writing data to Google Sheets...")

for row in sales_data:
    worksheet.append_row(row)

print(f"Done! You can find your sheet at: {sh.url}")
```

# OpenAI API

```python
import os
from google.colab import userdata
from openai import OpenAI

# 1. Pull the key from your Colab Secrets
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

# 2. Initialize the client
client = OpenAI()

sheet_url = "https://docs.google.com/spreadsheets/d/1HzVoKTL5eY45GXjffh5jfDMuokje8J3tBuZVa9iByJc/export?format=csv"
import pandas as pd

sheet_url = "https://docs.google.com/spreadsheets/d/1HzVoKTL5eY45GXjffh5jfDMuokje8J3tBuZVa9iByJc/export?format=csv"

df = pd.read_csv(sheet_url)

#print(df.head())
sales_data = df.values.tolist()

#print(sales_data)


# 3. Use triple quotes for the multi-line message content
#Interpret this sales data set.
#Give me insights, trends, and recommendations based on this sales data
user_prompt = input("What would you like to ask about the data? ")

response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "user", "content": f"""
            {user_prompt}



            sales_data = {sales_data}
        """}
    ]
)

print(response.choices[0].message.content)
#print(response)
```
