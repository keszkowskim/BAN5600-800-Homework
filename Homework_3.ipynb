1 - [5pts] Write program with while loop which finds the followings for the given values between 150 to 1000.

a) How many odd numbers are between 150 and 1000 (include 150 and 1000)

b) Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)

c) Find the multiplication of odd numbers which are between 150 and 1000 (include 150
and 1000)

d) Find the sum of square of odd numbers which are between 150 and 1000 (include 150
and 1000)



```python
count = 0
sum_odd = 0
product = 1
sum_squares = 0

num = 150

while num <= 1000:
    if num % 2 == 0:
        num += 1
        continue  # skip even numbers

    count += 1
    sum_odd += num
    product *= num
    sum_squares += num ** 2

    num += 1

print("Count of odd numbers:", count)
print("Sum of odd numbers:", sum_odd)
print("Product of odd numbers:", product)
print("Sum of squares of odd numbers:", sum_squares)
```

3 - [1pts] Write a program so that for given M, it computes the sum s = sum from k=1 to M of (1/k).

(Hint: Use the while-loop . As we see in in the class, you can update s as s=s+k and you need to increase
the k in each step such that k=k+1).

```python
M = int(input("Enter a value for M: "))

k = 1
s = 0

while k <= M:
    s = s + (1 / k)
    k = k + 1

print("The sum is:", s)
```

4 - [1pts] Write a program which repeatedly reads numbers until the user enters “done”. Once “done” is
entered, print out the total, count, and average of the numbers. If the user enters anything other than a
number, detect their mistake using try and except and print an error message and skip to the next number.
For example:
Enter a number: 4
Enter a number: 5
Enter a number: bad data Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

```python
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == "done":
        break

    try:
        num = float(user_input)
        total += num
        count += 1
    except:
        print("Invalid input")
        continue

# Avoid division by zero
if count > 0:
    average = total / count
else:
    average = 0

print(total, count, average)
```

5 - [1pts] Write another program that prompts for a list of numbers as above (question 4) and at the end
prints out both the maximum and minimum of the numbers instead of the average.

```python
max_num = None
min_num = None

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == "done":
        break

    try:
        num = float(user_input)

        if max_num is None or num > max_num:
            max_num = num

        if min_num is None or num < min_num:
            min_num = num

    except:
        print("Invalid input")
        continue

print("Maximum:", max_num)
print("Minimum:", min_num)
```
