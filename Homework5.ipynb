**B) Tech With Tim — Return Values + Default Parameters**
-What’s the difference between printing a value inside a function vs returning a value from a function? (2–4 sentences). Printing a value inside a function displays the result to the screen but does not send it back to be used elsewhere in the program. Returning a value sends the result back to the part of the code that called the function, allowing it to be stored in a variable or used in further calculations.  It depends on the system design as to which works best.

```python
def calc_total(price, tax_rate=None):
    if tax_rate is None:
        tax_rate = 0.07
    return price + (price * tax_rate)

# using default tax rate
print(calc_total(100))

# using custom tax rate
print(calc_total(100, 0.10))
```

-In your own words, what is the difference between a local variable and a global variable? (2–4 sentences) A local variable is created inside a function and can only be used within that function. A global variable is defined outside of functions and can be accessed from anywhere in the program. Local variables help keep things contained and avoid conflicts, while global variables are useful when multiple parts of a program need the same data.
